```python
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
def f1(x,y) :
    return(0.,0.16*y)
def f2(x,y):
    return(0.85*x + 0.04*y , -0.04*x + 0.85*y + 1.60)
def f3(x,y):
    return(0.2*x - 0.26*y , 0.23*x + 0.22*y + 1.60)
def f4(x,y):
    return(-0.15*x + 0.28*y , 0.26*x + 0.24*y + 0.44)
func = [f1,f2,f3,f4]

points = 100000
x, y=0, 0
x_list = []
y_list = []
for i in range(points) :
    function = np.random.choice(func , p=[0.01 , 0.85 , 0.07 , 0.07])
    x, y = function(x, y)
    x_list.append(x)
    y_list.append(y)

#for plotting
plt.scatter(x_list , y_list ,s=0.02 ,color = 'green')
plt.show()
```

```python
width , height = 300 , 300
fern_img = np.zeros((width, height))
x, y =0, 0
for i in range(points) :
    function = np.random.choice(func , p=[0.01 , 0.85 , 0.07 , 0.07])
    x, y = function(x, y)
    ix , iy = int(width/2 + x*width/10) , int(y*height/12)
    fern_img[iy,ix] = 1
plt.imshow(fern_img[::-1,:] , cmap = cm.Greens)
plt.savefig("Barnsley_fern")
plt.show()
```
